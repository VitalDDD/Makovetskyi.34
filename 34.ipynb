```python
import plotly
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objs as go
import pandas as pd
import numpy as np

# Читаємо файл з даними про міста
df_city=pd.read_excel('City.xls')

# Дивимось перші записи даних 
df_city.head()
```

```python
# Друге завдання - залежності населення міста від його площі та середнього доходу від населлення

fig = make_subplots(rows=2, cols=1)

fig.add_scatter(x=df_city['Площа (км²)'], y=df_city['Населення'], row=1, col=1)

fig.add_scatter(x=df_city['Населення'], y=df_city['Середній дохід'], row=2, col=1)

fig.update_layout(title="Залежності населення міста від його площі та середнього доходу від населлення")
fig.update_xaxes(title='Площа (км²)', col=1, row=1)
fig.update_xaxes(title='Населення', col=1, row=2)
fig.update_yaxes(title='Населення', col=1, row=1)
fig.update_yaxes(title='Середній дохід', col=1, row=2)
```

```python
# Третє завдання - матриця кореляцій

df_corr = df_city.corr()
fig = go.Figure()
fig.add_trace(go.Heatmap(x=df_corr.columns, y=df_corr.index, z=df_corr, colorscale='Blues'))
```

```python
# Четверте завдання - звіт за допомогою Sweetviz
import sweetviz as sv
result = sv.analyze(df_city)
result.show_notebook()
```

```python
# Четверте завдання - звіт за допомогою Pandas Profiling

from pandas_profiling import ProfileReport
import ydata_profiling

profile = ProfileReport(df_city)
profile
```

Коротке порівняння Pandas-profiling та Sweetviz:

Pandas-profiling створює докладний звіт про профіль даних за допомогою одного рядка коду. У звіті міститься більшість елементів профілювання даних, включаючи описову статистику та метрики якості даних. Він дає уявлення про статистику для окремих характеристик даних, таких як розподіл, а також середнє, мінімальне та максимальне значення. Звіт дає уявлення про кореляції та взаємодії між змінними.

Sweetviz – це комплексний дашборд із привабливим дизайном, який охоплює значну частину необхідного аналізу профілювання даних. Ця бібліотека дає змогу порівнювати дві версії одного набору даних.
